```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy.stats import linregress
from pprint import pprint

# Impor the OpenWeatherMap API key
from api_keys import weather_api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

### Generate the Cities List by Using the `citipy` Library

```python
# Empty list for holding the latitude and longitude combinations
lat_lngs = []
latslist=[]
longslist=[]

# Empty list for holding the cities names
cities = []

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        latslist.append(lat_lng[0])
        longslist.append(lat_lng[1])

# Print the city count to confirm sufficient count (Number of cities in the list: 585)
print(f"Number of cities in the list: {len(cities)}")
# cities
city_list = pd.DataFrame({"City":cities, "Approx. Latitude":latslist, "Approx. Longitude":longslist})
city_list
```

```python
city_list["Max Temperature"] = ""
city_list["Humidity (%)"] = ""
city_list["Cloudiness"] = ""
city_list["Wind Speed"] = ""
city_list["Latitude"] = ""
city_list["Longitude"] = ""
city_list["Country"] = ""
city_list["Datetime (Local)"] = ""
```

```python
city_data = []
base_url = " https://api.openweathermap.org/data/2.5/weather?"
for i, city in enumerate(cities):
    params = {"q":city_list["City"][i], 
              "units":"metric", 
              "appid":weather_api_key}
    response = requests.get(base_url, params=params)
#     print(response.url)
    city_weather =  response.json()
#     print(city_weather.keys())
#     pprint(city_weather)
    try:
        city_list.loc[i,"Latitude"]  = city_weather["coord"]["lat"]
        city_list.loc[i,"Longitude"]= city_weather["coord"]["lon"]
        city_list.loc[i,"Max Temperature"]  =city_weather["main"]["temp_max"]
        city_list.loc[i,"Humidity (%)"] = city_weather["main"]["humidity"]
        city_list.loc[i,"Cloudiness"] = city_weather["clouds"]["all"]
        city_list.loc[i,"Wind Speed"] = city_weather["wind"]["speed"]
        city_list.loc[i,"Country"] = city_weather["sys"]["country"]
        city_list.loc[i,"Datetime (Local)"] = city_weather["dt"]
        
        
    except:
#         print("City not found. Skipping...")
        pass
        
city_list
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
# Define a function to create Linear Regression plots
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


```

```python
# Create a DataFrame with the Northern Hemisphere data (Latitude >= 0)
northern_hemi_df = city_data_df[city_data_df["Latitude"] >= 0]
# Display sample data
northern_hemi_df.head()
```

```python
# Create a DataFrame with the Southern Hemisphere data (Latitude < 0)
southern_hemi_df=city_data_df[city_data_df["Latitude"]<= 0]

# Display sample data
southern_hemi_df.head()
```

###  Temperature vs. Latitude Linear Regression Plot

```python
# Linear regression on Northern Hemisphere
plt.figure(figsize=(9,6))
x_values = northern_hemi_df['Latitude']
y_values = northern_hemi_df['Max Temperature']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))


plt.scatter(x_values,y_values, marker='o', color='royalblue', s=[70], edgecolors='black')
plt.plot(x_values,regress_values, 'darkred', linewidth=2)


plt.title('Northern Hemisphere - Max Temp vs. Latitude', fontsize=20)
plt.ylabel('Max Temperature', fontsize=16, color='black')
plt.xlabel('Latitude', fontsize=16, color='black')
plt.annotate(line_eq,(1, 1), fontsize=18, color='darkred')


plt.ylim(min(city_data_df['Max Temperature'])-10, max(city_data_df['Max Temperature'])+10)
plt.xlim(-3, max(city_data_df['Latitude'])+3)

plt.savefig('../output_data/City Latitude vs. MaxTemp_North.png')

# Show plot:
plt.show()
# Print r-squared value:
print(f'The r-squared is: {round(rvalue**2,2)}')
```

```python
# Linear regression on Southern Hemisphere
plt.figure(figsize=(9,6))
x_values = southern_hemi_df['Latitude']
y_values = southern_hemi_df['Max Temperature']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq2 = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))


plt.scatter(x_values,y_values, marker='o', color='royalblue', s=[70], edgecolors='black')
plt.plot(x_values,regress_values, 'darkred', linewidth=2)


plt.title('southern Hemisphere - Max Temp vs. Latitude', fontsize=20)
plt.ylabel('Max Temperature', fontsize=16, color='black')
plt.xlabel('Latitude', fontsize=16, color='black')
plt.annotate(line_eq2,(-25,10), fontsize=18, color='darkred')

plt.ylim(min(southern_hemi_df['Max Temperature'])-10, max(southern_hemi_df['Max Temperature'])+10)

plt.xlim(min(city_data_df['Latitude'])-3,+1)

plt.savefig('../output_data/City Latitude vs. MaxTemp_south.png')

# Show plot:
plt.show()
# Print r-squared value:
print(f'The r-squared is: {round(rvalue**2,2)}')
```

**Discussion about the linear relationship:
1-if we compare the r_squared of teh northen vs southern hemisphere, we can conclude that the further the city from equater the lower the maximum temperature.
(strong negative corrlation in northen hemisphere vs positive correlation in the southern one)
2-we relaize as well that the maximum temperture in the sohern hemsphere is higher than on the northen one.
